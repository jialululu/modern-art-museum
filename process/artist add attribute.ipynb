```python
with open('data/moma_artists.json', encoding = 'utf-8') as f:
    artist = json.load(f)
artist[0]
```

```python
len(artist)
```

```python
color[190]['Domain color'][0][1]
```

```python
for i in range(len(artist)):
    ID = artist[i]['ConstituentID']  #the ID in artworks is a list
#     print(ID)
    main_prob=[]
    main_color =[]
    
    for j in range(len(color)):
        if color[j]['ConstituentID'][0]==ID:
            main_prob.append(color[j]['Domain color'][0][1])  # the probability of the main color
            main_color.append(color[j]['Domain color'][0][0])  #the main color RGB 
        else:
            pass
        
#     print(main_prob)
#     print(main_color)
    
    if len(main_prob) != 0:  # if the artist ID is not in artworks
        idx = main_prob.index(max(main_prob))
        c = main_color[idx]
        artist[i]['Represent_color']=c
    else:
        artist[i]['Represent_color']=[0,0,0]
    
#     if i >1000:
#         break

    if i%100==0:
        print('finish {} artists'.format(i))
```
