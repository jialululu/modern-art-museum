```python
import pandas as pd
```

```python
data = pd.read_csv('SELECT_t____CTID_FROM_public_color_new_t.csv',names = ["image_id","artist_id","dominant_color","max_color","department","height","width"])
```

```python
data.head()
```

```python
newdata = data.groupby('artist_id').filter(lambda x: len(x)>9)
```

```python
newdata.groupby('artist_id').mean().shape
```

```python
image_set = set(newdata["image_id"].tolist())
```

```python
len(image_set)
```

```python
import json

with open('Artworks_s-1.json', 'r') as data_file:
    jsondata = json.load(data_file)

newjson = list()
for item in jsondata:
    if item['ObjectID'] in image_set:
        index = image_set.index(item['ObjectID'])  # object ID of the drawing 
        print(index)
        color_list = json.loads(newdata.iloc[index]['dominant_color'])
        print(newdata.iloc[index]['height'],newdata.iloc[index]['width'],newdata.iloc[index]['artist_id'])
        sum_pixel = sum([x[1] for x in color_list])
        for sub in color_list:
            sub[1] = sub[1]/sum_pixel
        item['Domain color'] = color_list
        item['Max Color']= json.loads(newdata.iloc[index]['max_color'])
        newjson.append(item)
        
with open('art10.json', 'w') as outfile:  
    json.dump(newjson, outfile,ensure_ascii=False, indent=4)        
```
