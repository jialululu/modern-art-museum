```python
for i in range(len(artist)):
    ID = artist[i]['ConstituentID']  #the ID in artworks is a list
#     print(ID)
    main_prob=[]
    main_color =[]
    
    for j in range(len(color)):
        if color[j]['ConstituentID'][0]==ID:
            main_prob.append(color[j]['Domain color'][0][1])  # the probability of the main color
            main_color.append(color[j]['Domain color'][0][0])  # the main color RGB 
            print('artist ID matched!', ID)
        else:
            pass
        
#     print(main_prob)
    print(main_color)
    
    if len(main_prob) == 1:  # if the artist ID is in artworks
#         main_prob.remove(max(main_prob))  # get the second largest color
        idx = main_prob.index(max(main_prob))
        c = main_color[idx]
        artist[i]['Represent_color']=c
    elif len(main_prob)>1:  # if there is more than one color in the list, get the second largest color 
        idx = main_prob.index(max(main_prob))
        main_prob.remove(main_prob[idx])
        main_color.remove(main_color[idx])
        idx_2 = main_prob.index(max(main_prob))
        c = main_color[idx_2]
    else:
        artist[i]['Represent_color']=[0,0,0]
    
    print(c)
    
    if i >100:
        break

    if i%100==0:
        print('finish {} artists'.format(i))
```

```python
new_artist = []

for i in range(len(artist)):
    prob=[]
    domain_c=[]
    
    ID = str(artist[i]['ConstituentID'])
#     print(ID)
    
    if ID in color_key.keys():
        temp = color_key[ID]  # return the list of all artworks of the artist
        for j in range(len(temp)):
            prob.append(temp[j]['Domain color'][0][1])
            domain_c.append(temp[j]['Domain color'][0][0])

        idx = prob.index(max(prob))
        if len(prob)==1:
            c = domain_c[idx]
            artist[i]['Represent_color']=c
            new_artist.append(artist[i])
        elif len(prob)>1:
            prob.remove(prob[idx])
            domain_c.remove(domain_c[idx])
            idx_2 = prob.index(max(prob))
            c=domain_c[idx_2]
            artist[i]['Represent_color']=c
            new_artist.append(artist[i])
#             print('more than one color')
        else:
            pass
    
    if i%100==0:
        print('finish {} artwork'.format(i))
        
#     if i>1000:
#         break
    
```
